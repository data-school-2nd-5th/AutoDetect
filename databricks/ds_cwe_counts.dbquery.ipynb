```sql
WITH counts AS (
  SELECT
    (SELECT COUNT(*) FROM 3dt2ndteam5.cwe.cwe_weaknesses) AS bronze_count,
    (SELECT COUNT(*) FROM 3dt2ndteam5.cwe.cwe_weaknesses_silver) AS silver_count
)
SELECT
  current_timestamp() AS generated_at_utc,
  from_utc_timestamp(current_timestamp(), 'Asia/Seoul') AS generated_at_kst,
  bronze_count,
  silver_count,
  bronze_count - silver_count AS count_gap
FROM counts;
```
